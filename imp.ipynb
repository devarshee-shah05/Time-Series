# Simple Moving Average(Smoothening the curves)
- Disadvantage:
1. Same importance to all data...but in time series importance should be given to  current/recent data

![](SMA.png)

```python
df['Open'].plot(figsize=(11,5))
```

```python
df['Open:10 days rolling'] = df['Open'].rolling(window=10,min_periods=1).mean()
```

```python
df[['Open','Open:10 days rolling']].plot()
```

# Cumulative Moving Average

![](CMA.png)

```python
# Using function named expanding to do its practical implementation
df['Open'].expanding().mean().plot(figsize=(10,5))
```

# Exponential Moving Average

((close-Prev EMA)*multiplier)+Prev EMA

Multiplier = 2/(EMA rolling window+1)

```python
# Let's keep smoothing factor - 0.1
df['EMA_0.1'] = df['Open'].ewm(alpha= 0.1, adjust= False).mean()
```
